## Central limit theorem (CLT)
**According to the CLT, the distribution of sample means is approximately gaussian or normal irrespective of the shape of the distribution of the underlying population**.

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Example1: Sampling from unifirm distribution
#### Generate normally distributed data

```python
N = 1000                                # Number of data points
data = np.random.randint(100, size = (N))
   
plt.figure(figsize = (6, 4)) 
sns.histplot(data, kde = True)

plt.xlabel('bins')
plt.ylabel(f'Frequency($\\mu$)')
plt.show()
```

```python
n = 100      # sample size

sample_means = [np.mean(np.random.choice(data, n)) for _ in np.arange(N)]

print(f'mean (data) = {round(np.mean(data), 4)}')
print(f'mean (sample means) = {round(np.mean(sample_means), 4)}')

plt.figure(figsize = (10, 5))
sns.histplot(sample_means, kde = True)
plt.title('Sample Mean Distribution')
plt.axvline(x = np.mean(sample_means), color = 'r')
plt.show()
```

### Example 2: Sampling from a geometric distribution
Draw **`N`** values from the **geometric** distribution, with the **probability** of an individual success equal to **`p`**...

```python
p = 0.02
N = 1000
data = np.random.geometric(p, N)

plt.figure(figsize = (15, 6))
sns.histplot(data, kde = True) 
plt.xlabel(f'Number of trials (bins)')
plt.title('Graph of X ~ G(0.02)')
plt.show()
```

```python
n = 100         # sample size

sample_means = [np.mean(np.random.choice(data, n)) for _ in np.arange(N)]

print(f'mean (data) = {round(np.mean(data), 4)}')
print(f'mean (sample means) = {round(np.mean(sample_means), 4)}')

plt.figure(figsize = (10, 5))
sns.histplot(sample_means, kde = True)
plt.title('Sample Mean Distribution')
plt.axvline(x = np.mean(sample_means), color = 'r')
plt.show()
```
